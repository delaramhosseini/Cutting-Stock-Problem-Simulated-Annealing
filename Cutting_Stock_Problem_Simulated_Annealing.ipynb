implementation of cutting stock problem with simulating annealing algorithm.

1) The class Simulated_Annealing is defined. The constructor initializes various instance variables, including the alpha (cooling rate), temperature, number of requests, list of requests, length of a roll, current answer (representing the cutting of rolls), current fitness (number of rolls used), list of neighbors (possible solutions), expected number of rolls, and maximum number of iterations.

2) The input function takes a file path as input and reads the necessary data from the file. The data includes the length of a roll, a list of requests, and the expected number of rolls. This data is stored in instance variables for later use.

3) The initialize function initializes the instance variables based on the input data. It sets the requests, length of a roll, expected number of rolls, number of requests, current answer (initialized with an initial index of requests), and current fitness (calculated using the fitness function).

4) The initial_index_of_request function generates an initial index for the requests. It initializes the index list with pairs of indexes (one from the beginning and one from the end of the list of requests). If the number of requests is odd, it excludes the last index from the initial indexes.

5) The fitness_function function calculates the fitness of a cutting (represented by the suggested cutting). It iterates over the indexes in the suggested cutting, incrementing the current sum of pieces. If the sum exceeds the length of a roll, a new roll is used, and the used_rolls counter is incremented. The method returns the total number of rolls used.

6) The creat_neighbours function generates a list of neighboring cuttings. It creates a deep copy of the current answer and randomly swaps two elements in the copy to create a new neighbor. This new neighbor is returned.

7) The next_state function generates a new state from the current state using the Simulated Annealing algorithm. It creates a new neighbor, calculates the fitness of the neighbor, and calculates the energy difference (deltaE) between the neighbor and the current state. If deltaE is less than zero, indicating an improvement in fitness, the current state is updated with the neighbor. Otherwise, a random number is generated, and if it is less than the probability defined by the temperature and deltaE, the current state is still updated with the neighbor.

8) The run function is the main loop of the Simulated Annealing algorithm. It iteratively generates new states using the next_state method until either the maximum number of iterations is reached or the fitness of the current state becomes equal to or less than the expected number of rolls. The temperature is decreased at each iteration according to the cooling rate alpha.

```python
import random
from copy import deepcopy 
import math
```

```python
class Simulated_Annealing:
    def __init__(self , maximum_itrations = 1000 , Temp = 500, alpha = 0.9999 ): 
        self.alpha = alpha
        self.temp = Temp
        self.number_of_requests = 0
        self.requests = []
        self.length_of_roll = 0
        self.current_answer = []
        self.current_fitness = 0
        self.neighbours = []
        self.expected_number_of_rolls = 0
        self.maximum_itrations=maximum_itrations


    def input(self , path):
        file = open(path, 'r')
        lines = file.readlines()
        length_of_roll = int(lines[0].split()[-1])
        requests = list(map(int,lines[3][:-1].split(', ')))
        requests.sort()
        expected_number_of_rolls =  int(lines[5].split()[-1])
        self.initialize(requests ,length_of_roll,expected_number_of_rolls)


    def initialize(self,requests ,length_of_roll,expected_number_of_rolls):
        self.requests = requests
        self.length_of_roll = length_of_roll
        self.expected_number_of_rolls = expected_number_of_rolls
        self.number_of_requests = len(requests)
        self.current_answer = self.initial_index_of_request()
        self.current_fitness = self.fitness_function(self.current_answer)
    
        
    def initial_index_of_request(self):
        number_of_requests = self.number_of_requests
        initial_indexes = []
        num = 0 
        while (num < number_of_requests/2):
            for reQuest in range(num, num+2):
                if (reQuest == num): 
                    j = reQuest
                    initial_indexes.append(j)
                else: 
                    j = (number_of_requests - 1) - (reQuest - 1)
                    initial_indexes.append(j)
                    break
            num += 1
        
        if (number_of_requests % 2 != 0): 
            return initial_indexes[:-1]
                
        return initial_indexes
            
        
    def fitness_function(self , suggested_cutting): 
        requests = self.requests
        used_rolls = 1
        current_sum_of_pieces = 0
        for index in suggested_cutting:
            current_sum_of_pieces+= requests[index]
            if (current_sum_of_pieces > self.length_of_roll):
                used_rolls +=1
                current_sum_of_pieces = requests[index]
            
        return used_rolls 
    


    def creat_neighbours(self):       
        current_neighbour =[]
        current_neighbour = deepcopy(self.current_answer)
        rand1 = random.randint(0,self.number_of_requests-1)
        rand2 = random.randint(0,self.number_of_requests-1)
        current_neighbour[rand1],current_neighbour[rand2]=current_neighbour[rand2],current_neighbour[rand1]
        return  current_neighbour
        
    
        
        
    def next_state(self ,itration ):
        current_neighbour = self.creat_neighbours()
        # print (current_neighbour)
        current_fitness = self.current_fitness
        neibour_fitness = self.fitness_function(current_neighbour)
        deltaE = neibour_fitness - current_fitness
        if deltaE < 0:
            self.current_fitness = neibour_fitness
            self.current_answer = current_neighbour
            print (f"in itration number {itration} best fitness is {current_fitness} with temp of {self.temp}")
            
        
        else:
            rand = random.random()
            if rand < math.exp(-1*deltaE/self.temp):
                self.current_fitness = neibour_fitness
                self.current_answer = current_neighbour
            
            
        
    def run(self):
        itration = 0
        while (itration < self.maximum_itrations and self.current_fitness > self.expected_number_of_rolls):
            itration +=1
            self.next_state(itration)
            self.temp *=self.alpha
        print (f"in itration number {itration} best fitness is {self.current_fitness} with temp of {self.temp}")
            
    
        
    def print_answer(self): 
        print("final answer is : ")
        requests = self.requests
        used_rolls = 1
        current_sum_of_pieces = 0
        current_cutting = []
        for index in self.current_answer:
            current_sum_of_pieces+= requests[index]
            current_cutting.append(index)
            if (current_sum_of_pieces > self.length_of_roll):
                print(f"{used_rolls} : {current_cutting[:-1]}")
                used_rolls +=1
                current_sum_of_pieces = requests[index]
                current_cutting=current_cutting[-1:]
        print(f"{used_rolls} : {current_cutting[:]}")
        
    
    
    def prnt(self):
        print (f"number_of_requests is : {self.number_of_requests}")
        print (f"requests is : {self.requests}")
        print (f"length_of_roll is : {self.length_of_roll }")
        print (f"current_answer is : {self.current_answer}")
        print (f"current_fitness is : {self.current_fitness }")
        print (f"neighbours is : {self.neighbours}")
        print (f"expected_nember_of_rolls is : {self.expected_number_of_rolls }")
```
